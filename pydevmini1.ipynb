```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 25000 # Choose any! We auto support RoPE Scaling internally!
dtype = None # None for auto detection. Float16 for Tesla T4, V100, Bfloat16 for Ampere+
load_in_4bit = False # Use 4bit quantization to reduce memory usage. Can be False.

# 4bit pre quantized models we support for 4x faster downloading + no OOMs.
fourbit_models = [
    "unsloth/Meta-Llama-3.1-8B-bnb-4bit",      # Llama-3.1 2x faster
    "unsloth/Meta-Llama-3.1-8B-Instruct-bnb-4bit",
    "unsloth/Meta-Llama-3.1-70B-bnb-4bit",
    "unsloth/Meta-Llama-3.1-405B-bnb-4bit",    # 4bit for 405b!
    "unsloth/Mistral-Small-Instruct-2409",     # Mistral 22b 2x faster!
    "unsloth/mistral-7b-instruct-v0.3-bnb-4bit",
    "unsloth/Phi-3.5-mini-instruct",           # Phi-3.5 2x faster!
    "unsloth/Phi-3-medium-4k-instruct",
    "unsloth/gemma-2-9b-bnb-4bit",
    "unsloth/gemma-2-27b-bnb-4bit",             # Gemma 2x faster!

    "unsloth/Llama-3.2-1B-bnb-4bit",           # NEW! Llama 3.2 models
    "unsloth/Llama-3.2-1B-Instruct-bnb-4bit",
    "unsloth/Llama-3.2-3B-bnb-4bit",
    "unsloth/Llama-3.2-3B-Instruct-bnb-4bit",

    "unsloth/Llama-3.3-70B-Instruct-bnb-4bit" # NEW! Llama 3.3 70B!
] # More models at https://huggingface.co/unsloth

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "bralynn/pydevmini1", # or choose "unsloth/Llama-3.2-1B-Instruct"
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
    # token = "hf_...", # use one if using gated models like meta-llama/Llama-2-7b-hf
)
```

```python

import gradio as gr
import torch
from transformers import TextStreamer
from threading import Thread
from queue import Queue
from typing import List, Tuple

# 3. Define a custom streamer to yield text for Gradio
# This is the key to making the generation stream into the Gradio UI
class GradioTextStreamer(TextStreamer):
    def __init__(self, tokenizer, skip_prompt=True, **kwargs):
        super().__init__(tokenizer, skip_prompt, **kwargs)
        self.token_queue = Queue()
        self.stop_signal = object() # A unique sentinel to signal the end

    def on_finalized_text(self, text: str, stream_end: bool = False):
        """Puts the new text onto the queue."""
        self.token_queue.put(text)
        if stream_end:
            self.token_queue.put(self.stop_signal)

    def __iter__(self):
        return self

    def __next__(self):
        """Allows us to iterate over the queue to get generated tokens."""
        value = self.token_queue.get()
        if value is self.stop_signal:
            raise StopIteration()
        else:
            return value

# 4. The main chat function that Gradio will call
def chat_with_unsloth(
    message: str,
    history: List[Tuple[str, str]],
    temperature: float,
    top_p: float,
    max_new_tokens: int
):
    """
    This function processes the user's message and chat history,
    formats it for the Unsloth model, and generates a streaming response.
    """
    # FastLanguageModel.for_inference(model) is already called in the notebook

    # Format the conversation history and the new message
    messages = []
    for user_msg, assistant_msg in history:
        messages.append({"role": "user", "content": user_msg})
        messages.append({"role": "assistant", "content": assistant_msg})
    messages.append({"role": "user", "content": message})

    # Apply the chat template
    prompt = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True,
    )

    inputs = tokenizer([prompt], return_tensors="pt").to("cuda")
    streamer = GradioTextStreamer(tokenizer, skip_prompt=True)

    # Define generation parameters in a dictionary
    generation_kwargs = dict(
        inputs,
        streamer=streamer,
        max_new_tokens=max_new_tokens,
        temperature=temperature,
        top_p=top_p,
        use_cache=True,
    )

    # Use a separate thread for generation to avoid blocking the Gradio UI
    # This is crucial for streaming in Gradio
    thread = Thread(target=model.generate, kwargs=generation_kwargs)
    thread.start()

    # Yield the generated tokens as they become available
    full_response = ""
    for new_text in streamer:
        full_response += new_text
        yield full_response

# 5. Build and launch the Gradio app
with gr.Blocks(theme=gr.themes.Soft(), title="Unsloth Finetuned Chat") as demo:
    gr.Markdown("# Chat with your Unsloth Fine-tuned Model")
    gr.Markdown("Enter a message and see the model's streaming response. Adjust generation parameters on the left.")

    with gr.Row():
        with gr.Column(scale=1):
            # Generation Parameters
            gr.Markdown("### Generation Parameters")
            temperature_slider = gr.Slider(minimum=0.0, maximum=2.0, value=0.7, step=0.1, label="Temperature")
            top_p_slider = gr.Slider(minimum=0.1, maximum=1.0, value=0.8, step=0.05, label="Top P")
            max_new_tokens_slider = gr.Slider(minimum=64, maximum=80192, value=1024, step=64, label="Max New Tokens")

        with gr.Column(scale=3):
            # Chat Interface
            gr.ChatInterface(
                fn=chat_with_unsloth,
                chatbot=gr.Chatbot(height=600, label="Unsloth Chat", show_copy_button=True),
                additional_inputs=[
                    temperature_slider,
                    top_p_slider,
                    max_new_tokens_slider
                ],
                autofocus=True,
                examples=[
                    ["Continue the fibonacci sequence: 1, 1, 2, 3, 5, 8,"],
                    ["Write a short story about a dragon who loves to bake."],
                    ["Explain the importance of bees to the ecosystem in simple terms."],
                ]
            )

# Launch the app. `share=True` is vital for it to work in Google Colab.
# `queue()` enables handling of multiple requests.
demo.queue().launch(share=True, debug=True)
```
